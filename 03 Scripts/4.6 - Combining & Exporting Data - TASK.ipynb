We have product_id column common to both dataframes, now is time to check the shape of df_prods to understand what kind of join is more appropriate.


```python
# Check df_prods shape

df_prods.shape
```

We chose to use the inner join and merge the 2 dataframes because the information would be useless if it doesnt exist in both dataframes, I would like though to try to see the result of the merging without overwriting the dataframes, ince i already have a \_merge column i should put in the "indicator" another name. Here I choose 'new_merge'.


```python
# test merging inner

pd.merge(
    df_ords_combined, df_prods, on=["product_id"], how="inner", indicator="new_merge"
)
```

```python
# test merging outer and check the values count of new_merge
pd.merge(
    df_ords_combined, df_prods, on=["product_id"], how="outer", indicator="new_merge"
)["new_merge"].value_counts()
```

I think i need to drop the old "\_merge" column from the last merge, as now it is useless for the purpose.
The 2 database are not perfectly mergeble but keeping those data would be useless, so we will choose to use a Inner merge


```python
# Dropping column '_merge' in the df

df_ords_combined = df_ords_combined.drop("_merge", axis=1)
```

```python
# Checking if it dropped

df_ords_combined.head()
```

Let's proceed with the merge.


```python
# merging into orders_products_combined

df_orders_products_combined = df_ords_combined.merge(
    df_prods, on=["product_id"], how="inner", indicator=True
)
```
